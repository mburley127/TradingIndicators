```python
### Import Stock Data for the Ticker List
tickers = ['CELH', 'LI', 'AMD', 'TSM', 'BA', 'RDDT', 'ONCT', 'PFE', 'STT', 'MSAI', 'WNS', 'TSLA', 'TNON', 'BTDR', 'RRGB', 'AVO', 'JPM', 'EL', 'CRDO', 'VC', 'BLK', 'APLD', 'LCID', 'ORCL', 'RIVN', 'RUN', 'QS', 'GME', 'CISS', 'ALLR', 'SCHW', 'TMO', 'ZM', 'NMRK', 'HD', 'PYPL', 'IBM', 'BABA', 'PINS', 'SPOT', 'XOM', 'CRSR', 'LMND', 'SE', 'JNJ', 'EXC', 'ABNB', 'ETSY', 'CVX', 'ACN', 'SNAP', 'NIO', 'SBUX', 'AAPL', 'MRO', 'CSCO', 'UPST', 'CRM', 'NKE', 'LYV', 'TDOC', 'ACRS', 'MCD', 'KO']
start_date = '2018-01-01'

stock_data = import_stock_data(tickers, start_date)
stock_data.tail()
```

```python
### Return RSI and Perform Filtering
# Function call to compute RSI
periods = 14
rsi_df, rsi_vals = calculate_rsi(stock_data, periods, tickers)
#rsi_df.tail()
# Set new filtered df to current rsi_df
rsi_df_filtered = rsi_df

# Filter based on constraints:
    # Below 40 - indicate oversold conditions which generates a potential BUY signal, expect the price to rebound or correct upward after being oversold from downward selling pressure
    # Above 60 - indicate overbought conditions which generates a potential SELL signal
for ticker in rsi_df.columns:
    if rsi_df_filtered[ticker].iloc[-1] >= 40 and rsi_df_filtered[ticker].iloc[-1] <= 60:
        del rsi_df_filtered[ticker]

# Print the filtered DataFrame (optional)
print(rsi_df_filtered.tail(5))
```

```python
### Return MACD and Perform Filtering
# Function call to compute MACD
macd_df, macd_vals = calculate_macd(tickers, stock_data)
macd_df.tail()

# Filter based on constraints ->
```

```python
### Return CCI and Perform Filtering
# Function call to compute CCI
cci_df, cci_vals = calculate_cci(stock_data, window = 14)
#rsi_df.tail()
# Set new filtered df to current rsi_df
cci_df_filtered = cci_df
# Filter based on constraints:
    # Below -100 - Oversold conditions, price has moved significantly below its average, suggesting a potential downtrend or extended period of weakness, generating a potential SELL signal
    # Above +100 - Overbought conditions, upward trend is beginning, generating a potential BUY signal
for ticker in cci_df.columns:
    if cci_df_filtered[ticker].iloc[-1] >= -100 and cci_df_filtered[ticker].iloc[-1] <= 100:
        del cci_df_filtered[ticker]

cci_df_filtered.tail()
```

```python
### Return ADX and Perform Filtering
# Function call to compute ADX
adx_df, adx_vals = adx_mult_tickers(stock_data, tickers, window = 14)
#rsi_df.tail()
# Set new filtered df to current rsi_df
adx_df_filtered = adx_df
# Filter based on constraints:
    # Below -100 - Oversold conditions, price has moved significantly below its average, suggesting a potential downtrend or extended period of weakness, generating a potential SELL signal
    # Above +100 - Overbought conditions, upward trend is beginning, generating a potential BUY signal
for ticker in adx_df.columns:
    if adx_df_filtered[ticker].iloc[-1] >= -100 and adx_df_filtered[ticker].iloc[-1] <= 100:
        del adx_df_filtered[ticker]

adx_df_filtered.tail()
```
